## Statistics

Let us try to understand some of the statistical measures.

```python
import math
import random
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning)
```

```python
from scipy.stats import skewnorm
```

### Skewness

```python
# creating a normal distribution 
dist = skewnorm.rvs(a=0, size=1000000)
```

```python
# mean of the dist
mean = dist.mean()
# median of the dist
median = np.median(dist)
# mode of the dist
mode = scipy.stats.mode(np.around(dist, decimals=2)).mode[0]

print('mean: ', mean, ' median: ', median, ' mode: ', mode)
```

```python
sns.distplot(dist, bins=200);
plt.axvline(x=mean, c='red', label='mean')
plt.axvline(x=median, c='blue', linestyle='--', label='median')
plt.axvline(x=mode, c='green', label='mode');
plt.legend(loc='upper right');
```

Ideally, for a pure normal distribution: mean, median and mode coincide.

```python
print('The skewness of the distribution is: ', scipy.stats.skew(dist))
```

```python
# creating a positive skewed distribution 
dist = skewnorm.rvs(a=5, size=100000)
```

```python
# mean of the dist
mean = dist.mean()
# median of the dist
median = np.median(dist)
# mode of the dist
mode = scipy.stats.mode(np.around(dist, decimals=2)).mode[0]

print('mean: ', mean, ' median: ', median, ' mode: ', mode)
```

```python
sns.distplot(dist, bins=120);
plt.axvline(x=mean, c='red', label='mean')
plt.axvline(x=median, c='blue', linestyle='--', label='median')
plt.axvline(x=mode, c='green', label='mode');
plt.legend(loc='upper right');
```

```python
print('The skewness of the distribution is: ', scipy.stats.skew(dist))
```

```python
# creating a negatively skewed distribution 
dist = skewnorm.rvs(a=-5, size=100000)
```

```python
# mean of the dist
mean = dist.mean()
# median of the dist
median = np.median(dist)
# mode of the dist
mode = scipy.stats.mode(np.around(dist, decimals=2)).mode[0]

print('mean: ', mean, ' median: ', median, ' mode: ', mode)
```

```python
print('The skewness of the distribution is: ', scipy.stats.skew(dist))
```

### Kurtosis

```python
from scipy.stats import kurtosis
```

```python
dist = scipy.stats.norm.rvs(size=1000000)
```

```python
sns.distplot(dist, bins=200);
```

```python
kurtosis(dist, fisher=True)
```

```python
kurtosis(dist, fisher=False)
```

If Fisher distribution, then 3 is subtracted from the kurtosis, therefore, a normal distribution will have the kurtosis of 0. If not Fisher distribution, normal distribution will have the kurtosis of value 3.

$Kurt[X] = E\Big[\Big(\frac{X - \mu}{\sigma}  \Big)^4   \Big]$ 

```python
from scipy.special import gamma
```

```python
# pdf of Pearson distribution of VI type
def pearson7(x, a, m):
    coeff = gamma(m) / (a * math.sqrt(math.pi) * gamma(m - 1/2))
    return coeff * pow((1 + pow(x/a, 2)), -m)
```

```python
dist = [pearson7(x, 1, 2.53) for x in np.linspace(-3, 3, 1000)]
```

```python
plt.scatter(np.linspace(-3, 3, 1000), dist, s=2);
plt.ylabel('pdf')
plt.xlabel('x');
```

```python
print('kurtosis: ', kurtosis(dist, fisher=True))
```

```python
dist = [pearson7(x, 1, 7) for x in np.linspace(-3, 3, 1000)]
```

```python
plt.scatter(np.linspace(-3, 3, 1000), dist, s=2);
plt.ylabel('pdf')
plt.xlabel('x');
```
